```python
# -*- coding: utf-8 -*-
"""
Created on Wed Nov 14 03:00:30 2018


"""

import csv
import numpy as np

# ** INISIALISASI ** 

"Memasukan data-data angka dari bobot & kriteria"
"ke dalam program."

# bpk = bobot prioritas kriteria
# bka = bobot kriteria alternatif
# alt = daftar alternatif (wisata kota Malang)

bpk = np.empty((5, 1))
bka = np.empty((17, 5))
alt = []
jenis_kriteria = []

with open('bobot_kriteria.csv', 'r') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    next(csv_reader)
    baris = 0
    for row in csv_reader:
        bpk[baris] = row[1]
        jenis_kriteria.append(row[2])
        baris = baris + 1

bpk = bpk.astype(int)    
csv_file.close()   

with open('alternatif_dan_kriteria.csv', 'r') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    next(csv_reader)    
    baris = 0
    for row in csv_reader:
        bka[baris,:] = row[1:]
        alt.append(row[0])
        baris = baris + 1

bka = bka.astype(int)
csv_file.close()
    
# ** Tahap 1 ** 
    
'1. Melakukan Normalisasi pada Bobot Prioritas Kriteria'
    
# bt = bobot ternormalisasi
    
size = np.shape(bpk)
bt   = np.empty(size)

total = 0
for idx, elm in enumerate(bpk):
    total = total + elm
    
for idx, elm in enumerate(bpk):
    bt[idx] = bpk[idx]/total

# ** Tahap 2 ** 
    
'2. Membuat Tabel Kriteria & Alternatif'
    
# ** Tahap 3 ** 
    
'3. Menghitung Vektor Si'

# si  = matriks yang berisi kumpulan nilai-nilai dari vektor kolom si
# msi = vektor baris yang berisi hasil multiply dari nilai-nilai yg dimiliki si
# sigma_msi = variabel yg berisi nilai sigma dari isi vektor msi

size = np.shape(bka)
si   = np.empty(size)
msi  = np.empty((size[0]))

'Mencari nilai matriks si'
for baris, elm1 in enumerate(bka):
    for kolom, elm2 in enumerate(elm1):
        if(jenis_kriteria[kolom] == "keuntungan"):
            si[baris, kolom] = pow(bka[baris,kolom], bt[kolom])
        else:
            si[baris, kolom] = pow(bka[baris,kolom], -bt[kolom])

'Mencari nilai vektor baris msi'            
for baris, elm1 in enumerate(bka):
    multiply = 1
    for kolom, elm2 in enumerate(elm1):
        multiply = multiply * si[baris, kolom]
    msi[baris] = multiply

'Mencari nilai variabel sigma_msi'
sigma_msi = 0
for idx, elm in enumerate(msi):
    sigma_msi = sigma_msi + elm

# ** Tahap 4 ** 
    
'4. Menghitung Vektor Vi'

# vi  = vektor msi yang setiap nilainya di-operasi bagikan dengan sigma_msi

size = np.shape(msi)
vi = np.empty(size)

'Mencari vektor vi. Vektor msi yang nilainya elemennya dibagi dengan sigma_msi'
for idx, elm in enumerate(msi):
    vi[idx] = elm / sigma_msi

# ** Tahap 5 ** 
    
'5. Urutkan hasil dari tahap 4'

# Langkah terakhir adalah mengurutkan nilai elemen dari vektor vi
# Urutan dimulai dari terbesar hingga yang terkecil
# Nilai terbesar merupakan tempat wisata yang direkomendasikan oleh SPK

Rekomendasi = [x for _,x in sorted(zip(vi,alt), reverse = 1)]


"""                                                        """"""
    Mulai dari sini, barisan prosedur untuk tampilan tahapan
    
"""                                                        """"""
def tahapan1():
    print("Menampilkan Bobot Prioritas Kriteria")
    for elm in bpk:
        print(elm)
        
    print("Menampilkan Bobot Prioritas Kriteria yang Ternormalisasi")
    for elm in bt:
        print(elm)

def tahapan3():
    print("Menampilkan Matriks Si")
    print(si)
    print("Menampilkan Vektor Multiply Si")
    for elm in msi:
        print("",elm)
    print("Menampilkan nilai variabel Sigma Multiply Si")
    print("",sigma_msi)
    
def tahapan4():
    print("Menampilkan Matriks Vi")
    for elm in vi:
        print("",elm)

def tahapan5():
    print("Menampilkan Hasil Perankingan")
    for idx, elm in enumerate(Rekomendasi):
        print("[",'{:27}'.format(elm),"]")
        
"""                                                        """"""
                Program Utama Jalan di sini
    
"""                                                        """"""
ulang = 1
while(ulang):
    print("\nLihat Proses Tahapan")
    print("a. Melakukan Normalisasi pada Bobot Prioritas Kriteria.")
    print("b. Menghitung Vektor Si, Vektor Multiply Si, dan variabel Sigma Multiply.")
    print("c. Menghitung Vektor Vi.")
    print("d. Melihat Hasil Akhir Perankingan.")
    pilihan = input("Pilih proses : ")
    
    if(pilihan == "a"):
        tahapan1()
    elif(pilihan == "b"):
        tahapan3()
    elif(pilihan == "c"):
        tahapan4()
    elif(pilihan == "d"):
        tahapan5()
    else:
        ulang = 0
        
    jeda = input()
```
